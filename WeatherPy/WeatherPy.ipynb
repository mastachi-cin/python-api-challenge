```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import json
from scipy.stats import linregress

# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(lat_range[0], lat_range[1], size=1500)
lngs = np.random.uniform(lng_range[0], lng_range[1], size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
# Save config information.
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "imperial"

# Build partial query URL
query_url = f"{url}appid={weather_api_key}&units={units}&q="

# set up lists to hold reponse info
city_ls = []
latitud_ls = []
longitud_ls = []
temperature_ls = []
humidity_ls = []
cloudiness_ls = []
wind_ls = []
country_ls = []
date_ls = []
city_num = 0

#cities = []
#cities.append("La Asuncion")

# Print begin of log
print("Beginning Data Retrieval")
print("-----------------------------")

# Loop through the list of cities and perform a request for data on each
for city in cities:
    
    # City number
    city_num += 1
    
    # Print actual record in process
    print(f"Processing Record {city_num} | {city}")
    response = requests.get(query_url + city).json()
    
    try:
        city_ls.append(response["name"])
        latitud_ls.append(response['coord']['lat'])
        longitud_ls.append(response['coord']['lon'])
        temperature_ls.append(response['main']['temp_max'])
        humidity_ls.append(response['main']['humidity'])
        cloudiness_ls.append(response['clouds']['all'])
        wind_ls.append(response['wind']['speed'])
        country_ls.append(response['sys']['country'])
        date_ls.append(response["dt"])
        
    except:  
        # Print message for exception
        print("City not found. Skipping...")  
        pass 
        
# Print end of log
print("-----------------------------")
print("Data Retrieval Complete")
print("-----------------------------")

#print(json.dumps(response, indent=4))

# create a data frame for weather data
weather_dict = {
    "city": city_ls,
    "lat": latitud_ls,
    "lng": longitud_ls,
    "max_temp": temperature_ls,
    "humidity": humidity_ls,
    "cloudiness": cloudiness_ls,
    "wind_speed": wind_ls,
    "country": country_ls,
    "date": date_ls
}

weather_data = pd.DataFrame(weather_dict)

        
```

## Latitude vs. Temperature Plot

```python
# Function that creates scatter plot based on parameters

def print_scatter(x_val, y_val, title_val, xlabel_val, ylabel_val, file_val):
    
    # Create scatter plot
    plt.scatter(x_val,y_val)
    
    # Get local time 
    local_time = time.strftime('%m/%d/%Y', time.localtime())
    
    # Plot title with date of analysis
    plt.title(f'{title_val} {local_time}')
    
    # X label
    plt.xlabel(xlabel_val)
    
    # Y label
    plt.ylabel(ylabel_val)
    
    # Save figure 
    plt.savefig(f"images/{file_val}.png")
    
    # Print plot to screen
    plt.show()
    return
```

```python
# Call to function that creates Scatter Plot for Latitude vs Temperature
print_scatter(weather_data['lat'], 
              weather_data['max_temp'], 
              "City Latitud vs Max Temperature", 
              'Latitude', 
              'Max Temperature (F)', 
              "lat_vs_temp")
    
```

## Linear Regression

```python
# OPTIONAL: Create a function to create Linear Regression plots

# Function that creates linear regression plots based on parameters

def print_lin_reg(x_val, y_val, title_val, xlabel_val, ylabel_val, file_val):
    
    # Perform a linear regression
    (slope, intercept, rvalue, pvalue, stderr) = linregress(x_val, y_val)

    # Get regression values
    regress_values = x_val * slope + intercept
    
    # Create line equation string
    line_eq = "y = " + str(round(slope,2)) + "x +" + str(round(intercept,2))
    
    # Create plot
    plt.scatter(x_val,y_val)
    plt.plot(x_val,regress_values,"r-")
    
    # Get local time 
    local_time = time.strftime('%m/%d/%Y', time.localtime())
    
    # Plot title with date of analysis
    plt.title(f'{title_val} {local_time}')
    
    # X label
    plt.xlabel(xlabel_val)
    
    # Y label
    plt.ylabel(ylabel_val)
    
    # Annotate the line equation
    plt.annotate(line_eq, (20,15), fontsize=15,color="red")
    
    # Save figure 
    plt.savefig(f"images/{file_val}.png")
    
    # Print r square value
    print(f"The r-squared is: {rvalue**2}")
    
    # Print plot to screen
    plt.show()
    return
```
